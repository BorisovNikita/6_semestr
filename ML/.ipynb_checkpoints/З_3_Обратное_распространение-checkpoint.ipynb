Как известно, небольшая ошибка в середине решения интеграла может очень далеко распространится и привести к отчислению. ## Занятие 3. Обратное распространение ошибки

Для начала, давайте рассмотрим подробно конкретный пример: 

нс_1.JPG

Обозначим:

$$X = [x_1; x_2]$$

$$H = [h_1; h_2; h_3]$$

$$W = (w_{ij}), i \in {1, 2}, j \in {1, 2, 3}$$

$$B = [b_1; b_2; b_3]$$

Думаю, смысл понятен из обозначений и схемы выше. А функцию активации просто обозначим $F(*)$.

Тогда вектор выхода $H$ выразится очень просто $H = F(X \cdot W + B)$.
Как видите, все просто и линейно, на всякий случай - матричное тождество выше значит следующее:

$$h_1 = F(x_1 \cdot w_{11} + x_2 \cdot w_{21} + b_1)$$
$$...$$
$$h_3 = F(x_1 \cdot w_{13} + x_2 \cdot w_{23} + b_1)$$.

Просто вот такие три функции, вы уже знаете, что они сложные - внутри $F(*)$ стоит линейная функция. В школе вас уже учили все это дифференцировать, заменяя буквы цифрами и оставляя одну переменную. В вузе вам написали буквы и рассказали про частные производные. Цепное правило в школе называлось - производная сложной функции. Вспомнили? Сейчас будем применять. НО сначала цель!

**Цель - найти градиент ошибки по всем параметрам модели**.

Сейчас мы более подробно поймем суть этой фразы... (Так, надеюсь, все помнят, что такое градиент и догадались по изученным ранее методам оптимизации, как он здесь оказался)



Максимально упрощаем понимание - пусть $T = X \cdot W + B$, это линейная функция, стоящая внутри $F(*)$. Таким образом:

$$H = F(T)$$

Не поленюсь и приведу картинку всего происходящего в виде дифференцируемого графа:

нс_2.JPG

Пусть $E$ - ошибка НС, а $Ω$ - вектор обучаемых параметров. Наша задача найти градиент $\dfrac{dE}{dΩ}$.

**Постановка задачи**

Дано $\frac{\partial E}{\partial H} = [\frac{\partial E}{\partial h_i}; \frac{\partial E}{\partial h_2}; \frac{\partial E}{\partial h_3}]$

(отметим, что теперь мы можем вычислить $ \frac{\partial E}{\partial T} $, так как у нас есть цепное правило и функция $T$).

Зная производную функции ошибок $ \frac{\partial E}{\partial T} $, можно без труда посчитать $ \frac{\partial E}{\partial W} $, $ \frac{\partial E}{\partial B} $ и наконец производную по входу - $ \frac{\partial E}{\partial X} $.

Теперь название - обратное распространение ошибки становится понятным. Достаточно провести пальцем от выхода НС к входу и посмотреть на рассуждения этого блока!

**3.1** Каковы размерности матриц из предыдущего блока? Какое общий принцип определяет эти размерности?

**3.2** Ну а теперь найдем подробно $ \frac{\partial E}{\partial T} $. Первая компонента вычисляется следующим образом:

$$ \frac{\partial E}{\partial t_1} =  \frac{\partial E}{\partial h_1} \cdot
\frac{\partial h_1}{\partial t_1}  = \frac{\partial E}{\partial h_1} \cdot 
F^{'}(t_1)$$

Совершенно аналогично :

$$\frac{\partial E}{\partial t_2} =  \frac{\partial E}{\partial h_2} \cdot 
F^{'}(t_2)$$

$$\frac{\partial E}{\partial t_3} =  \frac{\partial E}{\partial h_3} \cdot 
F^{'}(t_3)$$

Используя произведение Адамара (просто поэлементное умножение) результат можно записать кратко:

$$\frac{\partial E}{\partial T} =  \frac{\partial E}{\partial H} \circ 
F^{'}(T)$$

**3.3** Найти $\frac{\partial E}{\partial W}$.

**Ответ:** $\frac{\partial E}{\partial W} = x^T \cdot \frac{\partial E}{\partial T}$.

**3.4** Показать, что $\frac{\partial E}{\partial B} = \frac{\partial E}{\partial T} $.

Как можно интерпретировать такой результат?

**3.5**

Найти $\frac{\partial E}{\partial X}$

**Указание:**

$$\frac{\partial E}{\partial x_1} = \frac{\partial E}{\partial t_1} \cdot \frac{\partial t_1}{\partial x_1} + \frac{\partial E}{\partial t_2} \cdot \frac{\partial t_2}{\partial x_1} + \frac{\partial E}{\partial t_3} \cdot \frac{\partial t_3}{\partial x_1}  $$ 

Посмотрите на самую верхнюю схему и сообразите графическую интерпретацию этой производной.

**Ответ:** $\frac{\partial E}{\partial X} = \frac{\partial E}{\partial T} \cdot W^T $

# Обратное распространение вручную с помощью PyTorch

Начнем с создания набора данных для регрессии

```python
import torch
```

```python
from sklearn.datasets import make_regression

X, y, coef = make_regression(n_features=2, n_informative=2, coef=True, bias=0.5)
X = torch.from_numpy(X).to(dtype=torch.float32)
y = torch.from_numpy(y).to(dtype=torch.float32)
```

[Граф вычислений для этой задачи](https://i.ibb.co/2dhDxZx/photo-2021-02-15-17-18-04.jpg)

Реализуем функцию ошибок и ее производную

```python
class SquaredLoss:
    def forward(self, y_pred, y_true):
        return (y_pred - y_true) ** 2

    def backward(self, y_pred, y_true):
        self.dinputs = 2 * (y_pred - y_true)
```

Создаем новый нейрон с методом обратного распространения

```python
class Neuron:
    def __init__(self, n_inputs):
        self.weights = torch.randn(n_inputs)
        self.bias = torch.randn(1)

    def forward(self, inputs):
        self.inputs = inputs
        return torch.dot(self.weights, inputs) + self.bias

    def backward(self, dvalue):
        self.dweights = self.inputs.unsqueeze(-1) @ dvalue
        self.dinputs = dvalue * self.weights.unsqueeze(-1)
        self.dbias = dvalue
```

Приступим к обучению

```python
neuron = Neuron(2)
loss = SquaredLoss()
lr = 0.01
for epoch in range(100):
    losses = []
    for x_example, y_example in zip(X, y):
        y_pred = neuron.forward(x_example)
        l = loss.forward(y_pred, y_example)
        losses.append(l)

        loss.backward(y_pred, y_example)
        neuron.backward(loss.dinputs)

        neuron.weights -= lr * neuron.dweights
        neuron.bias -= lr * neuron.dbias
    if epoch % 20 == 0 : print(f'epoch {epoch} mean loss {sum(losses) / len(losses)}')
```
